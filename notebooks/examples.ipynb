```python
im = imageio.imread('./bbaug/notebooks/dog.jpg')
```

```python
bbs = BoundingBoxesOnImage(
    [
        BoundingBox(x1=42, x2=266, y1=71, y2=244),
        BoundingBox(x1=279, x2=341, y1=48, y2=71)
    ],
    im.shape
)
```

```python
ia.imshow(bbs.draw_on_image(im, size=2))
```

```python
aug = augmentations.rotate_bbox(10)
im_aug, bbs_aug = aug(image=im, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(im_aug, size=2))
```

```python
aug = augmentations.brightness(10)
im_aug, bbs_aug = aug(image=im, bounding_boxes=bbs)
ia.imshow(bbs_aug.draw_on_image(im_aug, size=2))
```
